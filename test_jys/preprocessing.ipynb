```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = 'VP' # freesolv / esol / lipo / logvp / VP
df = pd.read_csv(r'..\data\VP.csv')
df
```

```python
if dataset == 'esol':
    y = 'logp'
elif dataset == 'freesolv':
    y = 'freesolv'
elif dataset == 'lipo':
    y = 'lipo'
elif dataset == 'logvp':
    y = 'logvp'
elif dataset == 'VP':
    y = 'logvp'
elif dataset == 'cherry':
    y = 'logvp'
```

```python
plt.figure(figsize = (8,6))
ax = sns.histplot(df[y], color = 'darkgray', edgecolor='white', bins=20,  alpha = 1, stat='density')
sns.kdeplot(df[y], ax=ax, linewidth=1.5, color='black')
plt.xlabel('')
plt.ylabel('')
```

```python
plt.figure(figsize=(8, 2))
sns.boxplot(x=df[y], color='lightgray')
plt.xlabel('')
plt.show()
```

```python
Q1 = np.percentile(df[y], 25)
Q3 = np.percentile(df[y], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
lower_bound
```

```python
df[df[y] < lower_bound]
```
